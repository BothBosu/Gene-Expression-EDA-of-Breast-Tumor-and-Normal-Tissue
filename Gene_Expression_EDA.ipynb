```python
import pandas as pd

# Count the occurrences of 'yes' and 'no' in the metastasis column
metastasis_counts = dat_long['metastasis'].value_counts()

# Total number of genes
total_genes = 20245

# Calculate proportions
yes_count = metastasis_counts.get('yes', 0)
no_count = metastasis_counts.get('no', 0)

yes_proportion = yes_count / total_genes
no_proportion = no_count / total_genes

# Output the results
print(f"Count of 'yes': {yes_count}, Proportion: {yes_proportion:.4f}")
print(f"Count of 'no': {no_count}, Proportion: {no_proportion:.4f}")
```

```python
# Step 2: Calculate total FPKM for each gene
total_fpkm = dat_long.groupby('gene')['FPKM'].sum().reset_index()

# Step 3: Identify top 5 genes based on total FPKM
top_genes = total_fpkm.sort_values(by='FPKM', ascending=False).head(5)['gene']

# Step 4: Compute average FPKM for top genes in tumor and normal tissues
# Filter data for top genes
top_genes_data = dat_long[dat_long['gene'].isin(top_genes)]

# Create a column to distinguish between tumor and normal tissue
top_genes_data['TissueType'] = top_genes_data['tissue'].apply(
    lambda x: 'Tumor' if 'tumor' in x.lower() else 'Normal'
)

# Calculate average FPKM
average_fpkm = top_genes_data.groupby(['gene', 'TissueType'])['FPKM'].mean().reset_index()

# Step 5: Plot the bar chart with specified requirements
import matplotlib.pyplot as plt
import numpy as np

# Get the list of genes
genes = average_fpkm['gene'].unique()

# Set the positions and width for the bars
x = np.arange(len(genes))  # the label locations
width = 0.35  # the width of the bars

# Create a pivot table to align data properly
pivot_table = average_fpkm.pivot(index='gene', columns='TissueType', values='FPKM').fillna(0)

# Ensure the columns are in the desired order: Tumor first, then Normal
pivot_table = pivot_table[['Tumor', 'Normal']]

# Get the average FPKM values
tumor_means = pivot_table['Tumor']
normal_means = pivot_table['Normal']

# Initialize the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the bars: Tumor bars on the left
rects1 = ax.bar(x - width/2, tumor_means, width, label='Tumor', color='lightcoral')
rects2 = ax.bar(x + width/2, normal_means, width, label='Normal', color='lightgreen')

# Add text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Gene')
ax.set_ylabel('Average FPKM (log scale)')
ax.set_title('Average FPKM Values for Top 5 Genes in Tumor and Normal Tissues')
ax.set_xticks(x)
ax.set_xticklabels(genes)
ax.legend()

# Set y-axis to log scale
ax.set_yscale('log')

# Remove grid lines
ax.grid(False)

# Function to add labels above bars
def autolabel(rects):
    """Attach a text label above each bar displaying its height"""
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            # Format the label for log scale
            label = f'{height:.2f}'
            ax.annotate(label,
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 5),  # vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=8)

# Apply the labels to both sets of bars
autolabel(rects1)
autolabel(rects2)

# Adjust layout to prevent clipping of tick-labels
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

# List of genes you want to analyze
genes_of_interest = ['AKT1', 'BRCA1', 'BRCA2', 'CDH1', 'ERBB2', 'ESR1', 'MKI67', 'MUC1', 'PIK3CA', 'PTEN', 'TP53']

# Filter the DataFrame to include only the rows with the specified genes
filtered_df = dat_long[dat_long['gene'].isin(genes_of_interest)]

# Split the data into metastasis yes and no
metastasis_yes_df = filtered_df[filtered_df['metastasis'] == 'yes']
metastasis_no_df = filtered_df[filtered_df['metastasis'] == 'no']

# Initialize a list to store the results
results = []

# Perform Mann-Whitney U test for each gene
for gene in genes_of_interest:
    # Extract FPKM values for the current gene from both metastasis yes and no
    yes_fpkm = metastasis_yes_df[metastasis_yes_df['gene'] == gene]['FPKM']
    no_fpkm = metastasis_no_df[metastasis_no_df['gene'] == gene]['FPKM']

    # Perform Mann-Whitney U test
    stat, p_value = mannwhitneyu(yes_fpkm, no_fpkm, alternative='two-sided')

    # Append the results to the list
    results.append({'Gene': gene, 'Mann-Whitney U Statistic': stat, 'p-value': p_value})

# Convert the results list into a DataFrame
mann_whitney_results = pd.DataFrame(results)

# Create a bar plot for the p-values of each gene
plt.figure(figsize=(10, 6))
sns.barplot(x='Gene', y='p-value', data=mann_whitney_results, palette='viridis')

# Add horizontal line to indicate the 0.05 significance level
plt.axhline(0.05, color='red', linestyle='--', label='Significance level (0.05)')

# Add plot title and labels
plt.title('Mann-Whitney U Test p-values for 11 Genes (Metastasis Yes vs No)', fontsize=14)
plt.xlabel('Gene', fontsize=12)
plt.ylabel('p-value', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Assuming you have the dataframe 'dat_long' in Python
# Filter the dataframe for the gene 'ATP5H'
filtered_data = dat_long[dat_long['gene'] == 'BRCA1']

# Group by 'tissue' and 'metastasis', and calculate the mean FPKM for each group
grouped_data = filtered_data.groupby(['tissue', 'metastasis'])['FPKM'].mean().reset_index()

# Define a custom color palette to match the tissue types and metastasis status
custom_palette = {
    ('breast tumor', 'no'): '#FF9999',    # Light red for non-metastasis in tumor
    ('breast tumor', 'yes'): '#DDA0DD',   # Light purple for metastasis in tumor
    ('normal breast tissue', 'no'): '#99FF99',  # Light green for non-metastasis in normal
    ('normal breast tissue', 'yes'): '#008000'  # Heavy green for metastasis in normal
}

# Create the plot
plt.figure(figsize=(8, 6))

# Since we have two tissues, we can set x positions accordingly
tissues = ['breast tumor', 'normal breast tissue']
x_positions = [0, 1]

# Bar width and offsets for metastasis status
bar_width = 0.4
offsets = [-bar_width/2, bar_width/2]  # Positions for 'no' and 'yes' metastasis

# For each tissue, plot the bars for 'no' and 'yes' metastasis
for i, tissue in enumerate(tissues):
    for j, metastasis in enumerate(['no', 'yes']):
        data_row = grouped_data[(grouped_data['tissue'] == tissue) & (grouped_data['metastasis'] == metastasis)]
        if not data_row.empty:
            FPKM_value = data_row['FPKM'].values[0]
            color = custom_palette[(tissue, metastasis)]
            plt.bar(i + offsets[j], FPKM_value, width=bar_width, color=color, edgecolor=color)
        else:
            # Handle cases where there is no data (optional)
            pass

# Set x-ticks and labels
plt.xticks(x_positions, tissues)
plt.xlabel('Tissue Type')
plt.ylabel('Average FPKM')
plt.title('Average FPKM for BRCA1 by Tissue and Metastasis Status')

# Create custom legend
legend_elements = [
    Patch(facecolor='#FF9999', edgecolor='#FF9999', label='Breast Tumor (Non-Metastasis)'),
    Patch(facecolor='#DDA0DD', edgecolor='#DDA0DD', label='Breast Tumor (Metastasis)'),
    Patch(facecolor='#99FF99', edgecolor='#99FF99', label='Normal Tissue (Non-Metastasis)'),
    Patch(facecolor='#008000', edgecolor='#008000', label='Normal Tissue (Metastasis)')
]
plt.legend(handles=legend_elements, title='Legend', loc='upper right')

# Remove the black lines (edges) on the bars
ax = plt.gca()
for bar in ax.patches:
    bar.set_edgecolor(bar.get_facecolor())

# Show plot
plt.tight_layout()
plt.show()
```

```python
import duckdb
import pandas as pd
from scipy import stats
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming dat_long is your full DataFrame

# Step 1: Create an in-memory DuckDB connection
con = duckdb.connect(database=':memory:')

# Load the full data into DuckDB (no sampling)
con.execute("CREATE TABLE gene_data AS SELECT * FROM dat_long")

# Step 2: Query the full data to calculate mean expression per gene and tissue type
query = """
    SELECT
        gene,
        tissue,
        AVG(FPKM) AS avg_FPKM
    FROM gene_data
    GROUP BY gene, tissue
"""
avg_expression_df = con.execute(query).fetchdf()

# Step 3: Perform t-tests and differential expression analysis
diff_exp_results = []

# Get the unique genes from the full data
genes = avg_expression_df['gene'].unique()

for gene in genes:
    # Filter data for the current gene
    gene_data = avg_expression_df[avg_expression_df['gene'] == gene]

    # Separate into tumor and normal tissues, dropping any None or NaN values
    tumor_group = gene_data[gene_data['tissue'] == 'breast tumor']['avg_FPKM'].dropna()
    normal_group = gene_data[gene_data['tissue'] == 'normal']['avg_FPKM'].dropna()

    # Perform t-test only if both groups have valid data
    if len(tumor_group) > 0 and len(normal_group) > 0:
        t_stat, p_value = stats.ttest_ind(tumor_group, normal_group, equal_var=False)
        # Calculate log2 fold change
        log2_fc = np.log2(tumor_group.mean() / normal_group.mean()) if normal_group.mean() > 0 else np.nan
    else:
        t_stat, p_value, log2_fc = None, None, None  # Set None if one of the groups is empty

    # Append results
    diff_exp_results.append({
        'gene': gene,
        'log2_fc': log2_fc,
        't_stat': t_stat,
        'p_value': p_value
    })

# Step 4: Convert the results into a DataFrame
diff_exp_df = pd.DataFrame(diff_exp_results)

# Step 5: Drop any rows with missing or None p-values
valid_p_values = diff_exp_df['p_value'].dropna()

# Step 6: Check if there are valid p-values before performing multiple testing correction
if len(valid_p_values) > 0:
    # Apply multiple testing correction using Benjamini-Hochberg
    _, corrected_p_values, _, _ = multipletests(valid_p_values, method='fdr_bh')
    diff_exp_df.loc[valid_p_values.index, 'adjusted_p_value'] = corrected_p_values
else:
    print("No valid p-values for multiple testing correction.")

# Step 7: Filter for significant genes (e.g., adjusted p-value < 0.05)
if 'adjusted_p_value' in diff_exp_df.columns:
    significant_genes = diff_exp_df[diff_exp_df['adjusted_p_value'] < 0.05]
    print("Significant Differentially Expressed Genes:")
    print(significant_genes)
else:
    print("No significant genes found.")

# Step 8: Plot the results (Volcano plot)
if 'adjusted_p_value' in diff_exp_df.columns:
    diff_exp_df['-log10(p_value)'] = -np.log10(diff_exp_df['p_value'])
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=diff_exp_df, x='log2_fc', y='-log10(p_value)', hue='adjusted_p_value', palette='coolwarm', legend=False)
    plt.title('Volcano Plot of Differential Gene Expression')
    plt.xlabel('log2 Fold Change')
    plt.ylabel('-log10(p-value)')
    plt.axhline(y=-np.log10(0.05), color='r', linestyle='--', label='Significance Threshold')
    plt.axvline(x=1, color='g', linestyle='--', label='Fold Change Threshold (+/- 1)')
    plt.axvline(x=-1, color='g', linestyle='--')
    plt.legend()
    plt.show()
```

```python
import duckdb
import pandas as pd
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from a CSV (or any file type DuckDB supports) into dat_long_duck
# Adjust the file path and name according to your data location
con = duckdb.connect()
con.register('dat_long_duck', dat_long)

# Get a list of all unique genes in the dataset using DuckDB
all_genes = con.execute("SELECT DISTINCT gene FROM dat_long_duck").fetchdf()['gene'].values

# Split the data into normal and cancer (tumor) tissue using DuckDB
normal_tissue_df = con.execute("SELECT * FROM dat_long_duck WHERE tissue = 'normal breast tissue'").df()
cancer_tissue_df = con.execute("SELECT * FROM dat_long_duck WHERE tissue = 'breast tumor'").df()

# Initialize a list to store the results
results = []

# Perform Mann-Whitney U test for each gene
for gene in all_genes:
    # Extract FPKM values for the current gene from both normal and tumor tissues
    normal_fpkm = normal_tissue_df[normal_tissue_df['gene'] == gene]['FPKM']
    cancer_fpkm = cancer_tissue_df[cancer_tissue_df['gene'] == gene]['FPKM']

    # Perform Mann-Whitney U test
    if len(normal_fpkm) > 0 and len(cancer_fpkm) > 0:  # Check if there is data for both groups
        stat, p_value = mannwhitneyu(normal_fpkm, cancer_fpkm, alternative='two-sided')

        # Append the results to the list
        results.append({'Gene': gene, 'Mann-Whitney U Statistic': stat, 'p-value': p_value})

# Convert the results list into a DataFrame
mann_whitney_results = pd.DataFrame(results)

# Filter to display only genes with p-value < 0.05 (significant level)
significant_genes = mann_whitney_results[mann_whitney_results['p-value'] < 0.05]

# Create a bar plot for the p-values of significant genes
plt.figure(figsize=(12, 8))
sns.barplot(x='Gene', y='p-value', data=significant_genes, palette='viridis')

# Add horizontal line to indicate the 0.05 significance level
plt.axhline(0.05, color='red', linestyle='--', label='Significance level (0.05)')

# Add plot title and labels
plt.title('Mann-Whitney U Test p-values for Significant Genes (Normal vs Tumor)', fontsize=16)
plt.xlabel('Gene', fontsize=14)
plt.ylabel('p-value', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()

# Display the significant results
print(significant_genes)
```
